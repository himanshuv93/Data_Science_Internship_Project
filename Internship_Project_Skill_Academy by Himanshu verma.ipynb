### Importing necessary libraries to build our model

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.linear_model import Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.ensemble import ExtraTreesRegressor
```

```python
res = {'Model':[],'R2Score':[]}

def eval_model(model,ytest,ypred):
    r2s=r2_score(ytest,ypred)
    print('R2Score',r2s)
    
    res['Model'].append(model)
    res['R2Score'].append(r2s)
    
```

```python
X=df_am.drop('Sales',axis=1)
y=df_am['Sales']
```

```python
print(type(X))
print(type(y))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
print('Training data X',X_train.shape)
print('Training Data y',y_train.shape)
print('Test data X',X_test.shape)
print('Test data y',y_test.shape)
```

```python
df_am.columns
```

```python
df_am.info()
```

### One Hot Encodeing(ohe)

```python
# Applying one hot encoding
ohe = OneHotEncoder()
ohe.fit(X[['City','Country','Region','Segment','Ship Mode','State','Product Name','Category','Sub-Category']])
```

```python
column_trans = make_column_transformer((OneHotEncoder(categories=ohe.categories_),['City','Country','Region','Segment','Ship Mode','State','Product Name','Category','Sub-Category']),
                                       remainder='passthrough')
```

```python
column_trans
```

### Building Regreesion Model

#### 1) Linear Regression model

```python
lr = LinearRegression()
```

```python
pipe=make_pipeline(column_trans,lr)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_lr = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_lr)
```

```python
eval_model(lr,y_test,y_pred_lr)
```

#### 2) Lasso Regression Model

```python
las = Lasso()
```

```python
pipe=make_pipeline(column_trans,las)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_las = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_las)
```

```python
eval_model(las,y_test,y_pred_las)
```

#### 3) Random Forest Model

```python
rf = RandomForestRegressor(n_estimators=100, max_depth=10)
```

```python
pipe=make_pipeline(column_trans,rf)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_rf = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_rf)
```

```python
eval_model(rf,y_test,y_pred_rf)
```

#### 4) Applying Bagging on Random Forest model 

```python
from sklearn.ensemble import BaggingRegressor
```

```python
bag_rf = BaggingRegressor(base_estimator=rf, n_estimators=10,random_state=42)
```

```python
pipe=make_pipeline(column_trans,bag_rf)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_bag_rf = pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_bag_rf)
```

```python
eval_model(bag_rf,y_test,y_pred_bag_rf)
```

#### 4) Applying AdaBoost on Random Forest

```python
from sklearn.ensemble import AdaBoostRegressor
```

```python
boost_rf = AdaBoostRegressor(base_estimator=rf, n_estimators=10, learning_rate=0.1, random_state=42)
```

```python
pipe=make_pipeline(column_trans,boost_rf)
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred_boost_rf= pipe.predict(X_test)
```

```python
r2_score(y_test,y_pred_boost_rf)
```

```python
eval_model(boost_rf,y_test,y_pred_boost_rf)
```
